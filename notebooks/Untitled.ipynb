```python
prompts = ["direct", "top-k", "cot", "multistep"]

def process_group(df):
    fig, axs = plt.subplots(2, 2, figsize=(6, 6),sharex=True, sharey=True, gridspec_kw={'hspace': 0.3})
    axes = axs.flatten()
    
    for i, row in df.iterrows():
        api = wandb.Api()
        artifact = api.artifact(f"uncertainty-reimplimentation-results/run-{row["ID"]}-table:v0")
        table = artifact.get("table").get_dataframe()
        prompt = row["prompt"]
    
        ax = axes[prompts.index(prompt)]
        plot_confidence_error(table["gpt_eval"], table["confidence"], ax=ax, title=prompt)
        
    fig.tight_layout()
    fig.savefig(f"images/calibration/{model}-{dataset}.png")
```
